```python
# Q5)


# Generate synthetic dataset
np.random.seed(42)  # For reproducibility
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = x1 + np.random.normal(0, 0.1, n_samples)  # x2 is correlated with x1
x3 = np.random.rand(n_samples)
x4 = x1 + np.random.normal(0, 0.2, n_samples)  # x4 is also correlated with x1

# Create a DataFrame
data = pd.DataFrame({'Feature 1': x1, 'Feature 2': x2, 'Feature 3': x3, 'Feature 4': x4})

# Compute the correlation matrix
corr_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Add title
plt.title('Correlation Matrix Heatmap')

# Show plot
plt.show()
```

```python
# Q2)


# Set random seed for reproducibility
np.random.seed(15)

# Generate synthetic data
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a violin plot
fig = px.violin(df, y='Score', x='Grade',
                 box=True,
                 points='all',  # Show all points
                 title='Violin Plot of Scores by Grade',
                 labels={'Score': 'Score', 'Grade': 'Grade'})

# Show the plot
fig.show()
```

```python
# Q3)

# Set random seed for reproducibility
np.random.seed(20)

# Generate synthetic data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Aggregate sales data by Month and Day
heatmap_data = df.groupby(['Month', 'Day'])['Sales'].sum().reset_index()

# Create a heatmap
fig = px.density_heatmap(heatmap_data,
                         x='Day',
                         y='Month',
                         z='Sales',
                         color_continuous_scale='Viridis',
                         title='Sales Heatmap by Month and Day',
                         labels={'Day': 'Day', 'Month': 'Month', 'Sales': 'Total Sales'})

# Show the plot
fig.show()
```

```python
# Q4)

import plotly.graph_objects as go

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x.flatten(), y=y.flatten(), colorscale='Viridis')])

# Update layout
fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
                  scene=dict(xaxis_title='X Axis', yaxis_title='Y Axis', zaxis_title='Z Axis'),
                  autosize=True)

# Show the plot
fig.show()
```

```python
# Q1)

# Output to notebook (remove this line if running in a standalone script)
output_notebook()

# Generate data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)  # Sine of x

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line renderer
p.line(x, y, line_width=2, color='blue', legend_label='sin(x)')

# Show the plot
show(p)
```

```python
# Q3)

from bokeh.models import ColumnDataSource

# Output to notebook (remove this line if running in a standalone script)
output_notebook()

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruits', y_axis_label='Counts',
           height=400, width=600)  # Use height and width

# Add a bar renderer
p.vbar(x='fruits', top='counts', source=source, width=0.9, color='blue')

# Show the plot
show(p)

```

```python
# Q4)



# Output to notebook (remove this line if running in a standalone script)
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title='Histogram of Random Data',
           x_axis_label='Value',
           y_axis_label='Frequency',
           height=400,
           width=600)

# Add a quad renderer for the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='blue', line_color='black')

# Show the plot
show(p)
```

```python
# Q2)


# Output to notebook (remove this line if running in a standalone script)
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100
x = np.random.rand(n) * 100  # Random x values
y = np.random.rand(n) * 100  # Random y values
sizes = np.random.rand(n) * 50 + 10  # Random sizes for markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)  # Random colors

# Create a DataFrame
data = {
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
}
df = pd.DataFrame(data)

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a Bokeh figure
p = figure(title="Scatter Plot with Varying Sizes and Colors",
           x_axis_label='X Axis',
           y_axis_label='Y Axis',
           height=400,
           width=600)

# Add a scatter renderer
p.scatter(x='x', y='y', size='sizes', color='colors', source=source, fill_alpha=0.6)

# Show the plot
show(p)
```

```python
# Q5)

from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource

# Output to notebook (remove this line if running in a standalone script)
output_notebook()

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Create a meshgrid
xx, yy = np.meshgrid(x, y)

# Flatten the arrays for plotting
xx_flat = xx.flatten()
yy_flat = yy.flatten()
data_flat = data_heatmap.flatten()

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=xx_flat, y=yy_flat, value=data_flat))

# Create a Bokeh figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           tools="hover", tooltips=[("Value", "@value")],
           width=400, height=400)

# Create a color mapper
mapper = linear_cmap(field_name='value', palette="Viridis256", low=min(data_flat), high=max(data_flat))

# Add a rect renderer for the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=source, fill_color=mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)

```
